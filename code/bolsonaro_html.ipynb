```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
navegador.get("http://www.biblioteca.presidencia.gov.br/presidencia/ex-presidentes/bolsonaro/discursos/discurso-do-presidente-da-republica-jair-bolsonaro-durante-abertura-do-debate-geral-da-74a-sessao-da-assembleia-geral-das-nacoes-unidas-agnu-nova-iorque-eua")
```

```python
# Buscando pelo id do site (usar menu inspecionar do navegador):

from selenium.webdriver.common.by import By
discurso_bolsonaro = navegador.find_element(By.ID, 'parent-fieldname-text')
discurso_bolsonaro.text
```

```python
# Texto do discurso
discurso_bolsonaro_onu = discurso_bolsonaro.text

# Criar um arquivo .txt e escrever o texto no arquivo
with open("discurso_bolsonaro.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write(discurso_bolsonaro_onu)
```
